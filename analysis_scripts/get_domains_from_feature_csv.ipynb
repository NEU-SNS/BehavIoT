```python
import os 
import sys
import csv
import pandas as pd

input_dir = '../data/idle-2021-test-filtered-std'
out_dir = 'apeirodic_domains'
if not os.path.exists(out_dir):
    os.system('mkdir -pv %s' % out_dir)


for dev_file in os.listdir(input_dir):
    if not dev_file.endswith('.csv'):
        continue
    device_name = dev_file.split('.')[0]
    print(device_name)
    data = pd.read_csv(os.path.join(input_dir, dev_file))
    hosts = data['hosts']
    
    value_counts = hosts.value_counts()

    # # Accessing the unique values
    # unique_values = value_counts.index

    # # Accessing the corresponding counts
    # counts = value_counts.values
    
    # Convert value_counts to a DataFrame and set column names
    value_counts_df = pd.DataFrame({'Domain': value_counts.index, 'Flow': value_counts.values})


    # Save value_counts to a CSV file
    value_counts_df.to_csv(os.path.join(out_dir, dev_file), index=False)
    # print(unique_values, counts)
```
