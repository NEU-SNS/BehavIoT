```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# load dataset
periodic_data_file = 'periodic_data/destination_party/_overall.csv'
periodic_data = pd.read_csv(periodic_data_file)
print(periodic_data.columns)
# tips = sns.load_dataset("tips")

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# set the figure size
plt.figure(figsize=(14, 14))

# # top bar -> sum all values(smoker=No and smoker=Yes) to find y position of the bars
data = periodic_data[['device', 'periodic models']] # .sum().reset_index()
# print(data['device'])

# bar chart 1 -> top bars (group of 'smoker=No')
bar1 = sns.barplot(x="device",  y="periodic models", data=data, color='darkblue')



periodic_data['sum'] = periodic_data['first party'] + periodic_data['support party']
bar3 = sns.barplot(x="device", y="sum", data=periodic_data, estimator=sum, ci=None,  color='royalblue')

# bottom bar ->  take only smoker=Yes values from the data
data2 = periodic_data[['device', 'first party']]
# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar2 = sns.barplot(x="device", y="first party", data=data2, estimator=sum, ci=None,  color='lightblue')



# add legend
top_bar = mpatches.Patch(color='darkblue', label='third party')
mid_bar = mpatches.Patch(color='royalblue', label='support party')
bottom_bar = mpatches.Patch(color='lightblue', label='first party')
plt.legend(handles=[top_bar, mid_bar, bottom_bar])

# show the graph
plt.show()
```

```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# load dataset
periodic_data_file = 'periodic_data/destination_party/_overall.csv'
periodic_data = pd.read_csv(periodic_data_file)
print(periodic_data.columns)
periodic_data = periodic_data[periodic_data['periodic models'] != 0]
# set the figure size
plt.figure(figsize=(14, 14))


# # top bar -> sum all values(smoker=No and smoker=Yes) to find y position of the bars
data = periodic_data[['device', 'periodic models']] # .sum().reset_index()
periodic_data['sum'] = periodic_data['first party'] + periodic_data['support party']
data2 = periodic_data[['device', 'sum']]
# bottom bar ->  take only smoker=Yes values from the data
data3 = periodic_data[['device', 'first party']]



data['periodic models'] = [i / j * 100 for i,j in zip(data['periodic models'], periodic_data['periodic models'])]
# print(data['periodic models'])
data2['sum'] = [i / j * 100 for i,j in zip(data2['sum'], periodic_data['periodic models'])]

data3['first party'] = [i / j * 100 for i,j in zip(data3['first party'], periodic_data['periodic models'])]
# print(data3['first party'])

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar1 = sns.barplot(x="device",  y="periodic models", data=data, color='darkblue')
bar2 = sns.barplot(x="device", y="sum", data=data2, estimator=sum, ci=None,  color='royalblue')
bar3 = sns.barplot(x="device", y="first party", data=data3, estimator=sum, ci=None,  color='lightblue')

# add legend
top_bar = mpatches.Patch(color='darkblue', label='third party')
mid_bar = mpatches.Patch(color='royalblue', label='support party')
bottom_bar = mpatches.Patch(color='lightblue', label='first party')
plt.legend(handles=[top_bar, mid_bar, bottom_bar])

# show the graph
plt.show()
```
