```python
import sys
import os
import pandas as pd 
import csv
base_dir = '~'
```

```python
# Get periodic_data stat numbers for tables: 
input_dir = '%s/imc_analysis/periodic_data/destination_party' % base_dir

header = ['protocol', 'domain', 'period', 'org', 'party']
device_dic = {}
output = []
for dev_file in os.listdir(input_dir):
    if dev_file.endswith('.csv') and not dev_file.startswith('_'):
        device_name = dev_file.split('.')[0]

        print(device_name)
        sum_period_models = 0
        
                
        df = pd.read_csv(os.path.join(input_dir, dev_file))

        # Count the unique values in the 'party' column
        party_counts = df['party'].value_counts()
        for i in party_counts:
            sum_period_models += i
        
        # print(party_counts)
        # print(party_counts[1])
        device_dic[device_name] = [device_name, sum_period_models, party_counts.get(1,0), party_counts.get(2,0), party_counts.get(3,0)]
        output.append([device_name, sum_period_models, party_counts.get(1,0), party_counts.get(2,0), party_counts.get(3,0)])

with open(os.path.join(input_dir, '_overall.csv' ), 'w') as file:
    header = ['device', 'periodic models', 'first party', 'support party', 'third party']
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(output)


```

```python
## Merge Apeirodic files

import pandas as pd

# Read the two CSV files into separate DataFrames
df1 = pd.read_csv('%s/imc_analysis/aperiodic_data/destination_party_idle/_overall.csv' % base_dir)
df2 = pd.read_csv('%s/imc_analysis/aperiodic_data/destination_party_routine/_overall.csv' % base_dir)

# Merge the DataFrames based on the 'device' column
merged_df = pd.merge(df1, df2, on='device', how='outer', suffixes=('_file1', '_file2'))
# Calculate the sum of 'first', 'support', and 'third' columns
merged_df['aperiodic domain count_file2'] = merged_df['aperiodic domain count_file2'].fillna(0)
merged_df['first party_file2'] = merged_df['first party_file2'].fillna(0)
merged_df['support party_file2'] = merged_df['support party_file2'].fillna(0)
merged_df['third party_file2'] = merged_df['third party_file2'].fillna(0)

merged_df['aperiodic domain count'] = merged_df['aperiodic domain count_file1'] + merged_df['aperiodic domain count_file2']
merged_df['first party'] = merged_df['first party_file1'] + merged_df['first party_file2']
merged_df['support party'] = merged_df['support party_file1'] + merged_df['support party_file2']
merged_df['third party'] = merged_df['third party_file1'] + merged_df['third party_file2']

merged_df = merged_df[['device', 'aperiodic domain count', 'first party', 'support party', 'third party']]
# Save the merged DataFrame to a new CSV file
out_dir = 'aperiodic_data/destination_party'
if not os.path.exists(out_dir):
    os.system('mkdir -pv %s' % out_dir)
merged_df.to_csv(os.path.join(out_dir, '_overall.csv'), index=False)
```

```python
import os, sys, json, csv

device_category = {
    'home auto': ["amazon-plug", "bulb1", "dlink-mov", "gosund-bulb1", "govee-led1", "keyco-air", "magichome-strip", "meross-dooropener", "nest-tstat", "philips-bulb", "ring-chime1", "smartlife-bulb", "t-wemo-plug", "thermopro-sensor", "tplink-bulb", "tplink-plug"],
    'camera': ["dlink-camera", "icsee-doorbell", "lefun-cam-wired", "microseven-camera", "ring-camera", "ring-doorbell", "tuya-camera", "ubell-doorbell", "wansview-cam-wired", "wyze-cam", "yi-camera"],
    'smart speaker': ["echodot", "echodot3a", "echodot3c", "echodot4b", "echoflex1", "echoplus", "echoshow5", "echospot", "google-home-mini", "google-nest-mini1", "homepod", "homepod-mini1"],
    'hub': ["aqara-hub", "ikea-hub", "smartthings-hub", "switchbot-hub", "t-philips-hub", "wink-hub2"],
    'appliance': ["brewer", "dishwasher", "fridge", "ikettle", "microwave", "sousvide"]
}
categories = device_category.keys()

periodic_file = 'periodic_data/destination_party/_overall.csv'
activity_file = 'activity_data/destination_party/_overall.csv'
aperiodic_file = 'aperiodic_data/destination_party/_overall.csv'

periodic_data = {}
with open(periodic_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        device = row['device']
        values = [row['first party'], row['support party'], row['third party']] 
        periodic_data[device] = values
# print(periodic_data)
activity_data = {}
with open(activity_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        device = row['device']
        values = [row['first party'], row['support party'], row['third party']] 
        activity_data[device] = values
# print(activity_data)
aperiodic_data = {}
with open(aperiodic_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        device = row['device']
        values = [row['first party'], row['support party'], row['third party']] 
        aperiodic_data[device] = values
# print(aperiodic_data)

output_dict = {}
output_dict['Periodic'] = {}
total = [0,0,0]
for device in periodic_data:
    for category in categories:
        if device in device_category[category]:
            if category not in output_dict['Periodic']:
                output_dict['Periodic'][category] = [0,0,0]
            for i in range(len(periodic_data[device])):
                output_dict['Periodic'][category][i] += int(periodic_data[device][i])
                total[i] += int(periodic_data[device][i])
output_dict['Periodic']['Total'] = total
print(len(activity_data))
output_dict['Activity'] = {}
total = [0,0,0]
for device in activity_data:
    for category in categories:
        if device in device_category[category]:
            if category not in output_dict['Activity']:
                output_dict['Activity'][category] = [0,0,0]
            for i in range(len(activity_data[device])):
                output_dict['Activity'][category][i] += int(activity_data[device][i])
                total[i] += int(activity_data[device][i])
output_dict['Activity']['Total'] = total


output_dict['Aperiodic'] = {}
total = [0,0,0]
for device in aperiodic_data:
    for category in categories:
        if device in device_category[category]:
            if category not in output_dict['Aperiodic']:
                output_dict['Aperiodic'][category] = [0,0,0]
            for i in range(len(aperiodic_data[device])):
                output_dict['Aperiodic'][category][i] += int(float(aperiodic_data[device][i]))
                total[i] += int(float(aperiodic_data[device][i]))
output_dict['Aperiodic']['Total'] = total

with open('per_category2.json', 'w') as f:
    json.dump(output_dict, f)
```
